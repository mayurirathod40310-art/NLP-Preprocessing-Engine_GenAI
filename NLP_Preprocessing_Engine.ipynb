# Task 1:

1.What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are considered different because of case sensitivity. "Love" usually comes at the start of a sentence or may be used as a proper noun, while "love" is a normal word. If we don’t handle this, the model may treat them as two separate words, which can create confusion. That’s why we usually convert everything to lowercase.



2.What happens if stopwords are not removed?

If stopwords are not removed, the text will contain many common words like "is", "the", "and" which don’t add much meaning. This makes the data noisy and can affect the performance of models because they might focus on these unnecessary words instead of important ones.


3.Provide two real-world scenarios where removing stopwords can be harmful.

1. Sentiment Analysis  
In sentences like "I am not happy", removing the stopword "not" changes the meaning to "I am happy", which gives the wrong sentiment.

3. Question Answering Systems  
In questions like "What is your name?", removing words like "what" and "is" may break the structure of the question and make it harder for the system to understand the intent.

4.Explain the difference between stemming and lemmatization.

Stemming is a technique that removes the ending of words to get the root form, often without considering grammar.  
Example: "running" → "run", "better" → "bett"
Lemmatization is a more advanced method that converts words to their base or dictionary form using proper linguistic rules.  
Example: "running" → "run", "better" → "good"
Lemmatization is more accurate than stemming but requires more processing.

# Task 2:

```python
import re

def preprocess_text(text):
    try:
        # Handle empty or invalid input
        if not isinstance(text, str) or text.strip() == "":
            return [], ""
        
        # Convert text to lowercase
        text = text.lower()
        
        # Remove URLs and email patterns
        text = re.sub(r'http\S+|www\S+|https\S+', '', text)
        text = re.sub(r'\S+@\S+', '', text)
        
        # Remove numbers
        text = re.sub(r'\d+', '', text)
        
        # Handle repeated characters (soooo → soo)
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)
        
        # Remove special characters and emojis
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()
        
        # Tokenization (split into words)
        tokens = text.split()
        
        # Remove very short words (<=2) except "no" and "not"
        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
        
        # Join tokens back into sentence
        cleaned_sentence = " ".join(tokens)
        
        return tokens, cleaned_sentence
    
    except Exception as e:
        # In case of unexpected error
        return [], ""


# Small test (optional but good practice)
print(preprocess_text("I have 2 dogs!!!"))
```

# Task 3 : Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Testing the function
for i, sentence in enumerate(test_sentences, 1):
    tokens, clean_sentence = preprocess_text(sentence)
    
    print(f"\n--- Sentence {i} ---")
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
```

# Task 4: Token Analytics

```python
# Task 4: Token Analytics

import numpy as np

print("\n====== Token Analytics ======")

for i, sentence in enumerate(test_sentences, 1):
    tokens, _ = preprocess_text(sentence)
    
    if len(tokens) == 0:
        print(f"\n--- Sentence {i} ---")
        print("No valid tokens found.")
        continue
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens])
    
    print(f"\n--- Sentence {i} ---")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
```

Analysis
- The sentence with the most noise was: "Win $$$ now!!! Limited offer!!!" because it contained symbols, repeated punctuation, and unnecessary characters.

- The sentence that retained the most meaningful tokens after cleaning was: "I am not happy with this" because it preserved important words like "not" which are useful for understanding sentiment.

# Task 5: Frequency Analysis

```python
from collections import Counter

# Collect all tokens from all sentences
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Count frequency
word_counts = Counter(all_tokens)

# Top 10 most frequent words
print("\nTop 10 Most Frequent Words:")
for word, count in word_counts.most_common(10):
    print(word, ":", count)

# Top 5 least frequent words
print("\nTop 5 Least Frequent Words:")
least_common = word_counts.most_common()[::-1][:5]

for word, count in least_common:
    print(word, ":", count)
```

# Task 6 : Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, sentence = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Testing the pipeline
output = full_pipeline(test_sentences)

print("\nTokens:")
print(output["tokens"])

print("\nClean Sentences:")
for i, sent in enumerate(output["clean_sentences"], 1):
    print(f"{i}. {sent}")
```

This function applies preprocessing to a list of texts and returns structured output

# Task 7 : Error Handling

```python
edge_cases = [
    "",                 # empty string
    "😂😂😂😂",         # only emojis
    "1234567890"        # only numbers
]

for i, case in enumerate(edge_cases, 1):
    tokens, sentence = preprocess_text(case)
    
    print(f"\n--- Case {i} ---")
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", sentence)
```

The preprocessing function correctly handles edge cases such as empty strings, emojis, and numeric inputs by returning empty tokens and an empty cleaned sentence. This ensures that the pipeline does not crash and remains robust for real-world noisy data.
